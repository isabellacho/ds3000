```python
import pandas as pd
df_cab = pd.read_csv('cab_rides.csv')
df_weather = pd.read_csv('weather.csv')
```

```python
def get_gmt_from_unix (unix_time):
    from datetime import datetime
    
    gmt = datetime.fromtimestamp(unix_time)
    unix = gmt.strftime('%B %d %Y')
    
    return unix
```

```python
import math
def get_gmt_from_unix2 (unix_time):
    from datetime import datetime
    unix_time = math.ceil(unix_time/1000)
    gmt = datetime.fromtimestamp(unix_time)
    unix = gmt.strftime('%B %d %Y %I:%M:%S %p')
    
    return unix
```

```python
date = df_weather['time_stamp'].map(get_gmt_from_unix)
df_weather['date'] = date
df_weather
```

```python
date_cab = df_cab['time_stamp'].map(get_gmt_from_unix2)
df_cab['date'] = date
location = df_cab['source']
df_cab['location'] = location
df_cab
```

```python
#Merge the DataFrames

df_merged = pd.merge(df_cab, df_weather, how='inner', left_index=True, 
                     right_index=True, suffixes=('', '_drop'))

#Drop the duplicate columns

df_merged.drop([col for col in df_merged.columns if 'drop' in col], 
               axis=1, inplace=True)

df_merged
```

```python
df_merged.to_csv('cab_weather_data.csv')
```
